# NUMPY

```python
import numpy as np
```

```python
# Задание 1
a = np.array([[1, 2, 3, 3, 1],
             [6, 8, 11, 10, 7]]).T

mean_a = a.mean(axis=0)
print(mean_a)
```

```python
# Задание 2
a_centered = a - mean_a
print(a_centered)
```

```python
# Задание 3
a_centered_sp = a_centered[0].dot(a_centered[1])
N, M = a.shape
a_cov = a_centered_sp / (N - 1)  
print(a_cov)
```

# PANDAS

```python
import pandas as pd
```

```python
# Задание 1
dict_authors = {
    "author_id": [1, 2, 3],
    "author_name": ['Тургенев', 'Чехов', 'Островский']
}
authors = pd.DataFrame(dict_authors)
authors
```

```python
dict_book = {
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий',
                    'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    "price": [450, 300, 350, 500, 450, 370, 290]
}
book = pd.DataFrame(dict_book)
book
```

```python
# Задание 2
authors_price = pd.merge(authors, book, on='author_id', how='inner')
authors_price
```

```python
# Задание 3
authors_price.sort_values(by="price")
top5 = authors_price.tail()
top5
```

```python
# Задание 4
groupby = authors_price.groupby("author_name")
authors_stat = groupby.agg(price_min=("price", "min"), price_max=("price", "max"), price_mean=("price", "mean"))
authors_stat
```

```python
# Задание 5
authors_price["cover"] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
book_info = pd.pivot_table(authors_price, values='price', index=['author_name'], columns=['cover'], aggfunc=np.sum)
book_info['мягкая'] = book_info['мягкая'].fillna(0)
book_info['твердая'] = book_info['твердая'].fillna(0)
print(book_info)
```

```python
book_info.to_pickle('book_info.pkl')
book_info2 = pd.read_pickle('book_info.pkl')
book_info.equals(book_info2)
```
